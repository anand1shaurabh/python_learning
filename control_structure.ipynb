This notebook briefly explain the control structure in Python.

1. Conditional flow: Execute the code based on the condition(s) specified. It can be used to test single or multiple, independent or nested conditions. Common examples are *if* , *if-else*, and multiple *if-elif-else*

```python
# generate a random interger
import random # import random module
a = random.randint(1,1000) #generate a random integer between 1 and 1000 (both limits included)

# simple if statement

if a % 3==0:
  print ("3 is a factor of", a)

# if/else statement
if a % 5 ==0:
  print("5 is a factor of", a)
else:
  print("5 is not a factor of", a)

# if-elif-else
if a>0:
  print(a, "is a positive number")
elif a<0:
  print(a, "is a negative number")
else:
  print(a,"is neither negative not positive")


```

2. *for* loop: This is used to iterate over an iterable object. Iterable is an object that can return its one element at a time, allowing the user to loop over all the elements. The most common examples of iterable objects in Python are *list*, *strings*, and *tuples*. Other examples include *dictionary*, *set*, and *generator* objects. (We will not cover those in this excercise)  

```python
# let us create a list
list1 = [1,2,3,4]

for num in list1:
  print(num*num)

#create new list based on the operation on a list. This can also be done through list comprehension. We will come to that later

newlist = [] # this creates an empty list. We will use it store mew elements.

for num in list1:
  new_num = num**0.5
  newlist.append(new_num)
print(newlist)

# for loop for string

string1 = "apple"

for character in string1:
  print(character.upper()) #convert each character in string to upper case


```

3. *While*: Execute a code block till specified condition evaluates to TRUE

```python
# generate a random interger
import random # import random module
b = random.randint(1,1000) #generate a random integer between 1 and 1000 (both limits included)

while b <5:
  print(b*b)
  
```

4. *Continue* and *Break*: These two keywords are used to gain control over *loop* function.

```python
# some example of continue
# let us create a list of numbers to iterate

list2 = [random.randrange(-1000,1000) for num in range(10)] # this creates a list of 10 random number between -1000 and 1000 with increment of 5

# We will use "break" to create a new list of consequetive positive numbers in list 2
list3=[] # define an empty list

# code block to get consequetive positive numbers from list2
for num in list2:
  if num<0:
    break # the execution will stop as soon as first negative number is encountered. Python will exit the "for" loop.
  list3.append(num)
if len(list3)>0:
  print("list of consequetive positive number:", list3) # this print list3 only if it is not empty
else:
    print("list does not have consequetive positive number")

# We will use "continue" to search for the even number in list2
list4=[]

for num in list2:
  if num%2!=0:
    continue # As long as the above condition is not true, the below code block will evaluate
  list4.append(num)
if len(list4)>0:
  print("list of even number:", list4)
else:
  print("no even number found")

```
