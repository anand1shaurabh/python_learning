```python
import tensorflow as tf
import numpy as np
```

```python
testimage = "DSC_0608.JPG"
```

```python
from tensorflow.keras.preprocessing import image
img = image.load_img(testimage, target_size=(224,224))
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
mobile = tf.keras.applications.mobilenet.MobileNet()
```

```python
resized_img = image.img_to_array(img)
final_img = np.expand_dims(resized_img, axis=0)
final_img = tf.keras.applications.mobilenet.preprocess_input(final_img)
```

```python
prediction = mobile.predict(final_img)
```

```python
from tensorflow.keras.applications import imagenet_utils
```

```python
result = imagenet_utils.decode_predictions(prediction)
```

```python
print(result)
```
